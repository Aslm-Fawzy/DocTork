# Jaundice Disease
**Try To Predict if Patient has Jaundice Disease  or Not Based on (Total Serum Billirubin 'TSB')**

## Creating Data

```python
import pandas as pd
import numpy  as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(15, 8)}, font_scale=1.3)
%matplotlib inline
```

```python
normal_TSB = np.round(np.random.uniform(0.1 ,1,10000),3)
normal_TSB_3 = np.round(np.random.uniform(0.5 ,1,15000),4)
normal_TSB_2 = np.round(np.random.uniform(0.1 ,.5,25000),4)
Abnormal_TSB = np.round(np.random.uniform(1.1,100,50000),3)
```

```python
TSB = np.concatenate((normal_TSB,normal_TSB_2,normal_TSB_3, Abnormal_TSB), axis = 0)
TSB
```

```python
TSB= TSB.reshape(TSB.shape[0],1)
TSB
```

```python
np.random.shuffle(TSB.flat)
TSB
```

```python
All_Data = pd.DataFrame(data = TSB , columns=['TSB'] )
All_Data
```

```python
All_Data.duplicated().sum() 
```

```python
All_Data['Gender'] = np.nan
All_Data['Output'] = np.nan
for i in range (0,len(All_Data)):
    row = All_Data.iloc[i]
    print('Row No' , i)
    gender_list = ['Male','Female']
    row.loc['Gender'] = np.random.choice(gender_list)
    #Handling Normal Patient
    #TSB Up to 1 
    if (row.loc['TSB'] <= 1) :
            row['Output'] = 'Normal' 
            All_Data.iloc[i] = row
            
    #Handling Patients have Jaundice 
    #TSB higher than 1  
    elif (row.loc['TSB'] > 1)  : 
            row['Output'] = 'Jaundice'
            All_Data.iloc[i] = row
```
