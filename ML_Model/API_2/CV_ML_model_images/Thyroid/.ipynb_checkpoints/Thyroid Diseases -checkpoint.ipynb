```python
for i in range (0,len(All_Data)):
    row = All_Data.iloc[i]
    print('Row No' , i)
    #Handling Normal TSH - Normal FT3 - Normal FT4
    if (row.loc['TSH'] in normal_TSH) and (row.loc['FT3'] in normal_FT3) and (row.loc['FT4'] in normal_FT4) :
            row['Output'] = 'Normal' 
            All_Data.iloc[i] = row
            
    #Handling Hypothyroid
    #TSH high - FT3 low - FT4 low
    
    elif (row.loc['TSH'] > 4.2) and (row.loc['FT3'] in normal_FT3) and (row.loc['FT4'] in normal_FT4) : 
            row['Output'] = 'Hypothyroid'
            All_Data.iloc[i] = row 
            
    elif (row.loc['TSH'] > 4.2) and (row.loc['FT3'] < 2) and (row.loc['FT4'] in normal_FT4) :
            row['Output'] = 'Hypothyroid'
            All_Data.iloc[i] = row
            
    elif (row.loc['TSH'] > 4.2) and (row.loc['FT3'] in normal_FT4) and (row.loc['FT4']  < 0.93) : 
            row['Output'] = 'Hypothyroid'
            All_Data.iloc[i] = row                 
            
    elif (row.loc['TSH'] > 4.2) and (row.loc['FT3'] < 2) and (row.loc['FT4']  < 0.93) : 
            row['Output'] = 'Hypothyroid'
            All_Data.iloc[i] = row            
            
    elif (row.loc['TSH'] in normal_TSH) and (row.loc['FT3'] < 2) and (row.loc['FT4'] in normal_FT4) :
            row['Output'] = 'Hypothyroid'
            All_Data.iloc[i] = row  
            
    elif (row.loc['TSH'] in normal_TSH) and (row.loc['FT3'] < 2) and (row.loc['FT4']  < 0.93) :
            row['Output'] = 'Hypothyroid'
            All_Data.iloc[i] = row    
            
    elif (row.loc['TSH'] in normal_TSH) and (row.loc['FT3'] in normal_FT3) and (row.loc['FT4']  < 0.93) :
            row['Output'] = 'Hypothyroid'
            All_Data.iloc[i] = row               
    
    
    #Handling Hypothyroid
    #TSH low - FT3 high - FT4 high        
            
    elif (row.loc['TSH'] in normal_TSH) and (row.loc['FT3'] in normal_FT3) and (row.loc['FT4']  > 1.7) :
            row['Output'] = 'HyperTyroid'
            All_Data.iloc[i] = row         
            
    elif (row.loc['TSH'] in normal_TSH) and (row.loc['FT3'] > 4.4) and (row.loc['FT4']  > 1.7) :
            row['Output'] = 'HyperTyroid'
            All_Data.iloc[i] = row          
            
    elif (row.loc['TSH'] < 0.27) and (row.loc['FT3'] > 4.4) and (row.loc['FT4']  > 1.7) :
            row['Output'] = 'HyperTyroid'
            All_Data.iloc[i] = row  
            
    elif (row.loc['TSH'] < 0.27) and (row.loc['FT3'] in normal_FT3) and (row.loc['FT4']  > 1.7) :
            row['Output'] = 'HyperTyroid'
            All_Data.iloc[i] = row  
            
    elif (row.loc['TSH'] < 0.27) and (row.loc['FT3'] > 4.4) and (row.loc['FT4']  in normal_FT4) :
            row['Output'] = 'HyperTyroid'
            All_Data.iloc[i] = row           
            
    elif (row.loc['TSH'] < 0.27) and (row.loc['FT3'] in normal_FT3) and (row.loc['FT4']  in normal_FT4) :
            row['Output'] = 'HyperTyroid'
            All_Data.iloc[i] = row  
            
    elif (row.loc['TSH'] in normal_TSH) and (row.loc['FT3'] > 4.4) and (row.loc['FT4']  in normal_FT4) :
            row['Output'] = 'HyperTyroid'
            All_Data.iloc[i] = row                
         
    else :
            row['Output'] = 'Other Thyroid Abnormalities'
            All_Data.iloc[i] = row    
```

```python
All_Data.drop('Gender',axis = 1 , inplace = True)
```

```python

x = All_Data.drop('Output', axis=1)
y = All_Data['Output']
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x,y,test_size=0.2 , 
                                                     stratify= y ,shuffle=True)
```

## Modelling and Evalution


### Using One Model





```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report,fbeta_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter = 300),
    "KNN" : KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting"  :  GradientBoostingClassifier(),
    "XGB": XGBClassifier(),
}
```

```python
for name, model in models.items():
    print(f'Training Model : {name} \n---------------------------------------------------------')
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f'Training Accuracy: {model.score(x_train, y_train)}')
    print(f'Testing Accuracy: {accuracy_score(y_test, y_pred)}')
    print(f'Testing Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
    #print(f'Testing Recall: {recall_score(y_test, y_pred)}')    
    #print(f'Testing Precesion: {precision_score(y_test, y_pred)}')
    #print(f'Testing F-1: {f1_score(y_test, y_pred)}')
    #print(f'Testing F-Beta: {fbeta_score(y_test, y_pred, beta=0.5)}')    
    print('-'*70)
```

```python
model = RandomForestClassifier(n_estimators=250 ,max_depth=25) #0.992336890353076 (214)   
#GradientBoosti_ngClassifier(n_estimators=200,max_depth=20) #0.992336890353076 (214)//////
```

```python
model.fit(x_train, y_train)
print(f'Training Accuracy: {model.score(x_train, y_train)}')
y_pred = model.predict(x_test)
print(f'Testing Accuracy: {accuracy_score(y_test, y_pred)}')
```
