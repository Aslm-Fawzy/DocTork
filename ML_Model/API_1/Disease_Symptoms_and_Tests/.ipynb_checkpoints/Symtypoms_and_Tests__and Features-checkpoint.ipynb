# Symtypoms and Tests
**Try to predict some diseases based on specific symtypoms and show the suitable lab test for this disease**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy  as np
%matplotlib inline
sns.set(rc={'figure.figsize':(15, 8)}, font_scale=1.3)
```

## Reading  


```python
df = pd.read_excel('/content/Diseases and Symptoyms.xlsx')
```

```python
df
```

## Preprocessing

```python
df['Symptoms'] = 0
```

```python
records = df.shape[0]
for i in range(records):
    values = df.iloc[i].values
    values = values.tolist()
    if 0 in values:
        df["Symptoms"][i] = values[1:values.index(0)]
    else:
        df["Symptoms"][i] = values[1:]
```

```python
column_values = df[['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4',
       'Symptom 5', 'Symptom 6', 'Symptom 7', 'Symptom 8']].values.ravel()
symps = pd.unique(column_values)
symps = symps.tolist()
symps = [i for i in symps if str(i) != "nan"]
symptoms = pd.DataFrame(columns = symps,index = df.index)
```

```python
symptoms["Symptoms"] = df["Symptoms"]
```

```python
for i in symps:
    symptoms[i] = symptoms.apply(lambda x:1 if i in x.Symptoms else 0, axis=1)
symptoms["Disease"] = df["Disease"]
symptoms = symptoms.drop("Symptoms",axis=1)
symptoms.head()
```

```python
list(symptoms.columns)
```

```python
x = symptoms.drop('Disease',axis=1 )
y = symptoms['Disease']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x,y,test_size=0.4 , 
                                                     stratify= y ,shuffle=True)
```

```python
x_train
```

## Modelling and Evalution


### Using One Model





```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report,fbeta_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter = 300),
    "KNN" : KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting"  :  GradientBoostingClassifier(n_estimators=200,max_depth=3),
}
```

```python
for name, model in models.items():
    print(f'Training Model : {name} \n---------------------------------------------------------')
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f'Training Accuracy: {model.score(x_train, y_train)}')
    print(f'Testing Accuracy: {accuracy_score(y_test, y_pred)}')
    print(f'Testing Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
    #print(f'Testing Recall: {recall_score(y_test, y_pred)}')    
    #print(f'Testing Precesion: {precision_score(y_test, y_pred)}')
    #print(f'Testing F-1: {f1_score(y_test, y_pred)}')
    #print(f'Testing F-Beta: {fbeta_score(y_test, y_pred, beta=0.5)}')    
    print('-'*70)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
print(f'Training Accuracy: {model.score(x_train, y_train)}')
y_pred = model.predict(x_test)
print(f'Testing Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
print(f'Testing Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
```

```python
df_cm = pd.DataFrame(confusion_matrix(y_test,y_pred), columns=np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = 'Actual Labels'
df_cm.columns.name = 'Predicted Labels'
df_cm
```

```python
sns.set(rc={'figure.figsize':(12, 8)}, font_scale=1.5)
plt.title(" Confusion Matrix", size=20)
sns.heatmap( df_cm,cmap='Blues', annot=True,annot_kws={"size": 16},fmt='',
            linewidths=1.3,cbar=False , linecolor = 'w')
```

### Ensemble Technique

```python
estimator = [] 
estimator.append(('Random Forset',RandomForestClassifier())) 
estimator.append(('KNN', KNeighborsClassifier()))  #10
estimator.append(('LR',   LogisticRegression()))
```

```python
estimator
```

```python
from sklearn.ensemble import VotingClassifier
hard_voting = VotingClassifier(estimators = estimator, voting ='hard') 
hard_voting.fit(x_train, y_train) 
print("Hard Voting Traninig :" , hard_voting.score(x_train,y_train)) 
```

```python
y_pred = hard_voting.predict(x_test)
score = accuracy_score(y_test, y_pred) 
print("Hard Voting Testing :" , score) 
print(f'Testing Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
```

```python
df_cm = pd.DataFrame(confusion_matrix(y_test,y_pred), columns=np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = 'Actual Labels'
df_cm.columns.name = 'Predicted Labels'
df_cm
```

```python
sns.set(rc={'figure.figsize':(12, 8)}, font_scale=1.5)
plt.title(" Confusion Matrix", size=20)
sns.heatmap( df_cm,cmap='Blues', annot=True,annot_kws={"size": 16},fmt='',
            linewidths=1.3,cbar=False , linecolor = 'w')
```
